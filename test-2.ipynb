# Limpieza del CSV

Ya con los datos en la base documental en Mongo procedemos a extraer el .csv. Mongo, al exportar los documentos a .csv, tiene un pequeño error
tratando la cadena "0" y lo saca con el valor de false.


```python
# Librerias
import pandas as pd
import numpy as np
```

```python
# Importar los datos
dataset = 'data/expropiese.csv'
df = pd.read_csv(dataset, sep=',')
df.head(2)
```

Coger solo los de un determinado tipo

```python
selected_df = df[df['target'] == 4]
selected_df
```

Para culminar almacenamos el __.csv__ resultante

```python
save_path = 'data/dbm-model-positive.csv'
selected_df.to_csv(path_or_buf=save_path, index=False, sep=';')
```
