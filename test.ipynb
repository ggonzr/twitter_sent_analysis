# Limpieza del CSV

Ya con los datos en la base documental en Mongo procedemos a extraer el .csv. Mongo, al exportar los documentos a .csv, tiene un pequeño error
tratando la cadena "0" y lo saca con el valor de false.


```python
# Librerias
import pandas as pd
import numpy as np
```

```python
# Importar los datos
dataset = 'data/dbm-2.csv'
df = pd.read_csv(dataset, sep=',')
df.head(2)
```

```python
parsed_df = df[['Polarity', 'Text', 'created_at', 'id', 'user']]
parsed_df
```

Cambiar los nombres de las columnas

```python
parsed_df = parsed_df.rename(columns={
    'Polarity': 'target',
    'Text': 'text',
    'id': 'ids',
    'created_at': 'date',
    'user': 'user'
})
```

```python
parsed_df
```

```python
# Descripcion
parsed_df.head(2)
```

```python
# Agrupando
#_df = parsed_df.groupby(by='target')
#ist(g_df.groups)
```

```python
# Positivos
positive_df = parsed_df[(parsed_df['target'] =='P') | (parsed_df['target'] =='FN')]
positive_df
```

```python
# Neutros
neu_df = parsed_df[parsed_df['target'] == 'NEU']
neu_df
```

```python
# Seleccionamos un numero de filas para no desbalancear el experimento
neu_2_df = neu_df.iloc[0:7200]
neu_2_df
```

```python
# Negativos
negative_df = parsed_df[parsed_df['target'] == 'N']
negative_df
```

```python
# Concatenar los subconjuntos
#all_df = pd.concat([positive_df, neu_2_df, negative_df])
all_df = pd.concat([positive_df, negative_df])
all_df
```

```python
# Procesar el target
def target(row):
    target = row
    if target == 'P' or target == 'FN':
        return 4
    elif target == 'NEU':
        return 2
    elif target == 'N':
        return 0
    
all_df['target'] = all_df['target'].apply(target)
```

```python
# Resultado
all_df
```

Para culminar almacenamos el __.csv__ resultante

```python
save_path = 'data/dbm-model-result-2.csv'
all_df.to_csv(path_or_buf=save_path, index=False, sep=';')
```
