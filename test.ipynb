# Limpieza del CSV

Ya con los datos en la base documental en Mongo procedemos a extraer el .csv. Mongo, al exportar los documentos a .csv, tiene un pequeño error
tratando la cadena "0" y lo saca con el valor de false.


```python
# Librerias
import pandas as pd
import numpy as np
import random
```

```python
# Importar los datos
dataset = 'data/dbm-r2.csv'
df = pd.read_csv(dataset, sep=';')
df.head(6)
```

```python
df['date'] = df['date'].apply(lambda row: '')
df
```

```python
save_path = 'data/dbm-r3.csv'
df.to_csv(path_or_buf=save_path, index=False, sep=';')
```

Cambiar los nombres de las columnas

```python
# Duplicados
b_df = df.text.drop_duplicates()
idx = b_df.index
df.loc[idx]
```

```python
# Seleccion aleatoria
inicio = int(random.uniform(0, 7000))
aumento = int(random.uniform(5000, 7000))
fin = inicio + aumento
```

```python
pos_df = df[df['target'] == 0]
s_p_df = pos_df.iloc[inicio:fin]
s_p_df
```

```python
# Seleccion aleatoria
inicio = int(random.uniform(0, 7000))
aumento = int(random.uniform(5000, 7000))
fin = inicio + aumento
```

```python
neg_df = df[df['target'] == 4]
s_n_df = neg_df.iloc[inicio:fin]
s_n_df
```

Para culminar almacenamos el __.csv__ resultante

```python
# Colocar el tag
def tag(row, tag='cnn'):
    return f'{row} @{tag}'
```

```python
s_p_df['text'] = s_p_df['text'].apply(tag)
s_n_df['text'] = s_n_df['text'].apply(tag)
```

```python
all_df = pd.concat([s_p_df, s_n_df])
save_path = 'data/dbm-r2.csv'
all_df.to_csv(path_or_buf=save_path, index=False, sep=';')
```
