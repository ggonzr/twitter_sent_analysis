# Limpieza del CSV

Ya con los datos en la base documental en Mongo procedemos a extraer el .csv. Mongo, al exportar los documentos a .csv, tiene un peque√±o error
tratando la cadena "0" y lo saca con el valor de false.


```python
# Librerias
import pandas as pd
import numpy as np
```

```python
# Importar los datos
dataset = 'data/dbm-parsed.csv'
df = pd.read_csv(dataset, sep=';')
df.head(2)
```

```python
parsed_df = df[['Sentiment', 'TweetText', 'TweetDatetime', 'TweetId', 'TwitterName']]
parsed_df
```

Cambiar los nombres de las columnas

```python
parsed_df = parsed_df.rename(columns={
    'Sentiment': 'target',
    'TweetText': 'text',
    'TweetId': 'ids',
    'TweetDatetime': 'date',
    'TwitterName': 'user'
})
```

```python
parsed_df
```

Para culminar almacenamos el __.csv__ resultante

```python
save_path = 'data/dbm-model.csv'
parsed_df.to_csv(path_or_buf=save_path, index=False, sep=';')
```
