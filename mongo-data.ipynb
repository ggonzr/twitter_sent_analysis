# Limpieza del CSV

Ya con los datos en la base documental en Mongo procedemos a extraer el .csv. Mongo, al exportar los documentos a .csv, tiene un pequeño error
tratando la cadena "0" y lo saca con el valor de false.


```python
# Librerias
import pandas as pd
import numpy as np
```

```python
# Importar los datos
dataset = 'data/dbm.csv'
df = pd.read_csv(dataset)
df.head(2)
```

Tenemos el siguiente error

```python
df[df['Sentiment'] == 'false']
```

```python
# Resumen
df.describe()
```

A continuación aplicamos una funcion sobre cada uno de los datos

```python
df['Sentiment'] = df['Sentiment'].apply(
    lambda row: 0 if row == 'false' else row,    
)
```

Nuevo DataFrame

```python
df.head(10)
```

Mostramos algunos datos

```python
df[df['Sentiment'] == 0]
```

Para culminar almacenamos el __.csv__ resultante

```python
save_path = 'data/dbm-parsed.csv'
df.to_csv(path_or_buf=save_path, index=False, sep=';')
```
